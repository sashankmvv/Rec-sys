```python
import pandas as pd
from scipy import sparse
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df=pd.read_csv("Time_invested.csv",index_col='Users')
items=pd.read_csv("warehouses.csv")
```

```python
def standardize(row):
    new_row = (row - row.mean())/(row.max()-row.min())
    return new_row

df_std = df.apply(standardize)
print(df_std)
df_std_not_null=df_std.fillna(0)
# sparse_df = sparse.csr_matrix(df_std_not_null.values)
# sparse_df
```

```python
corrMatrix=cosine_similarity(df_std_not_null)
corrMatrix
```

```python
corrMatrix = pd.DataFrame(corrMatrix,index=df.index,columns=df.index)
corrMatrix
```

```python
# Number of similar users
n = 3
# User similarity threashold
user_similarity_threshold = 0
picked_userid='91-060-5966'
corrMatrix.drop(index=picked_userid, inplace=True)

# Get top n similar users

similar_users = corrMatrix[corrMatrix[picked_userid]>user_similarity_threshold][picked_userid].sort_values(ascending=False)[:n]
# Print out top n similar users
print(f'The similar users for user {picked_userid} are', similar_users)
```

```python
df_std
```

```python
picked_userid_invested = df_std[df_std.index == picked_userid].dropna(axis=1, how='all')
picked_userid_invested
```

```python
similar_user_invested=df_std[df_std.index.isin(similar_users.index)].dropna(axis=1, how='all')
similar_user_invested
```

```python
similar_user_invested.drop(picked_userid_invested.columns,axis=1, inplace=True, errors='ignore')
```

```python
similar_user_invested #the warehouses where the similar users have invested but the picked user has not
```

```python
# A dictionary to store item scores
item_score = {}
# Loop through items
for i in similar_user_invested.columns:
  # Get the ratings for movie i
  warehouse_rating = similar_user_invested[i]
  # Create a variable to store the score
  total = 0
  # Create a variable to store the number of scores
  count = 0
  # Loop through similar users
  for u in similar_users.index:
    # If the warehouse has rating
    if pd.isna(warehouse_rating[u]) == False:
      # Score is the sum of user similarity score multiply by the warehouse rating
      score = similar_users[u] * warehouse_rating[u]
      # Add the score to the total score for the warehouse so far
      total += score
      # Add 1 to the count
      count +=1
  # Get the average score for the item
  item_score[i] = total / count
# Convert dictionary to pandas dataframe
item_score = pd.DataFrame(item_score.items(), columns=['warehouse', 'warehouse_score'])
    
# Sort the warehouses by score
ranked_item_score = item_score.sort_values(by='warehouse_score', ascending=False)
# Select top m warehouses
m = 10
ranked_item_score.head(m)
```
